```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from sklearn.naive_bayes import GaussianNB

from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('wine.data', names=['class', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium', 'Total phenols', 'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins', 'Color intensity', 'Hue', 'OD280/OD315 of diluted wines', 'Proline'])
```

```python
target = 'class'
```

```python
# Problema de clasificación: hacer categórico el atributo target: class

df[target]=pd.Categorical(df[target])
```

```python
# Clasificador k-NN, con validación simple, k=5.

# X, Y

x = df.drop(['class'], axis=1)
y = df['class']
```

```python
RANDOM_SEED = 30

x_train, x_test, y_train, y_test = train_test_split(
  x, y, test_size=0.2, random_state=RANDOM_SEED)

print('Dimensiones de X,Y dataset de entrenamiento: {dimX}, {dimY}.'.format(dimX=x_train.shape, dimY=y_train.shape))
print('Dimensiones de X,Y dataset de validación: {dimX}, {dimY}.'.format(dimX=x_test.shape, dimY=y_test.shape))

clf = GaussianNB()

clf.fit(x_train, y_train)

# Predicción para datos de validación
y_pred = clf.predict(x_test)

# Tasa de aciertos para dataset de validación
acc_score = accuracy_score(y_test, y_pred)
print('Tasa de aciertos en dataset de validación: {acc_sc}'.format(acc_sc=acc_score))

# Predicción para dataset de entrenamiento
y_pred_train = clf.predict(x_train)

acc_score_train = accuracy_score(y_train, y_pred_train)
print('Tasa de aciertos en dataset de entrenamiento: {acc_sc}'.format(acc_sc=acc_score_train))
```
