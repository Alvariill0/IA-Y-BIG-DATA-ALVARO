## Creando matrices y visualizándolas como imagen

Las imágenes se codifican como matrices. En particular, las imágenes de intensidad o escala de grises se codifican como una matriz de dos dimensiones, donde cada número representa la intensidad de un pixel.

La función que nos permite visualizar matrices es imshow del modulo pyplot, que invocaremos como plt.imshow, y recibe la imagen como parámetro.

Debido a que pyplot intenta ajustar los colores automaticamente de una imagen, vamos a pasarle como parámetros también vmin=0,vmax=1. Si bien no los vimos antes, estos se llaman keyword parameters o parámetros por palabra clave de python

Tendremos que modificar el colormap con la sentencia  ```plt.rcParams['image.cmap'] = 'gray'```para que utilice escala de grises. 
Por defecto, Matplotlib aplica un colormap llamado viridis, que es un mapa de colores que va del amarillo al morado, pasando por una gama de colores intermedios. 


````python
# importamos el modulo pyplot, y lo llamamos plt
import matplotlib.pyplot as plt

#configuracion necesaria de pyplot para ver las imagenes en escala de grises
#plt.rcParams['image.cmap'] = 'gray'


# comando de Jupyter para que las imagenes se muestren automaticamente 
#%matplotlib inline

#tambien importamos numpy ya que lo usamos para crear y manipular matrices
import numpy as np
````

### 3. Explora cómo conseguir grises en tu figura

````python
# Si 0 = negro y 1 = blanco, para representar grises habrá que usar valores intermedios:
tamanio_matriz = (50,50)
plt.rcParams['image.cmap'] = 'gray'
# Usamos un valor intermedio:
gris = np.ones(tamanio_matriz) * 0.5
# Mostramos la imgen
plt.imshow(gris,vmin=0,vmax=1)
````

## 7. Combina 2 a 2 los tres canales anteriores. 
Usa como base una matriz de zeros de la misma estructura que la de la imagen. Para ello puedes usar la función de numpy ```zeros_like```

````python
# Cargar la imagen del astronauta
astronauta = data.astronaut()

# Obtener los canales de la imagen (R, G, B)
canal_rojo = astronauta[:,:,0]  # Canal rojo
canal_verde = astronauta[:,:,1]  # Canal verde

# Crear una nueva imagen combinada (rojo y verde)
# Ponemos el canal azul a 0 para que no influya en la imagen combinada
combinada = np.zeros_like(astronauta)

# Asignar el canal rojo y verde a la imagen combinada
combinada[:,:,0] = canal_rojo  # Canal rojo
combinada[:,:,1] = canal_verde  # Canal verde

# Mostrar la imagen combinada
plt.imshow(combinada)
plt.title("Canal Rojo y Verde Combinados")
plt.axis('off')  # Quitar ejes
plt.show()
````

````python
combinada_rb = np.zeros_like(astronauta)
combinada_rb[:,:,0] = astronauta[:,:,0]  # Rojo
combinada_rb[:,:,2] = astronauta[:,:,2]  # Azul
plt.imshow(combinada_rb)
plt.title("Combinación Rojo y Azul")
plt.show()
````

````python
combinada_gb = np.zeros_like(astronauta)
combinada_gb[:,:,1] = astronauta[:,:,1]  # Verde
combinada_gb[:,:,2] = astronauta[:,:,2]  # Azul
plt.imshow(combinada_gb)
plt.title("Combinación Verde y Azul")
plt.show()
````

## 8. Convierte la imagen a escala de grises

Una forma simple para hacer la conversión consiste en sacar el promedio de los canales y usar eso como la intensidad

````python
# Cargar la imagen
astronauta = data.astronaut()

h, w, c = astronauta.shape  # Obtener el tamaño de la imagen original

astronauta_gris = np.zeros((h, w))  # Crear una matriz donde generar la imagen en escala de grises

for i in range(h):
    for j in range(w):
        
        promedio = np.mean(astronauta[i, j, :3]) 
        astronauta_gris[i, j] = promedio  

plt.imshow(astronauta_gris, cmap='gray')  
plt.title("Imagen en Escala de Grises")
plt.show()
````

## 9. Haz un filtro sepia para la imagen. Para ello:

- Elige un colo narajita
- Superponlo a todos los pixeles
- Usa la función clip para no pasarte de 255

````python
# Cargar la imagen de la astronauta
astronauta = data.astronaut()

# Obtener las dimensiones de la imagen 
h, w, c = astronauta.shape

# Crear una copia de la imagen para aplicar el filtro
astronauta_sepia = np.copy(astronauta)

# Tono naranja
naranja = np.array([40, 30, 0])  

# Iterar sobre cada píxel de la imagen
for i in range(h):
    for j in range(w):
        # Obtener el valor RGB del píxel original
        r, g, b = astronauta[i, j]

        # Sumar el color
        astronauta_sepia[i, j] = np.clip([r + naranja[0], g + naranja[1], b + naranja[2]], 0, 255)

# Mostrar la imagen original y la imagen con filtro sepia
plt.figure(figsize=(10, 5))

# Imagen original
plt.subplot(1, 2, 1)
plt.imshow(astronauta)
plt.title("Imagen Original")
plt.axis('off')

# Imagen con filtro sepia
plt.subplot(1, 2, 2)
plt.imshow(astronauta_sepia)
plt.title("Filtro Sepia")
plt.axis('off')
````

## 10. Implementa una función que ajuste el contraste de la imagen usando un factor de contraste entre 0 y 2 a los tres canales de la imagen

````python
# Cargar la imagen 
astronauta = data.astronaut()

# Crear una copia de la imagen para modificarla
astronauta_contraste = astronauta.copy()

# Factor de contraste
factor_contraste = 0.5

# Recorrer todos los píxeles
for i in range(astronauta.shape[0]): 
    for j in range(astronauta.shape[1]):  
        for c in range(3):  #
            valor_pixel = astronauta[i, j, c]
            
            nuevo_valor = valor_pixel * factor_contraste

            nuevo_valor = np.clip(nuevo_valor, 0, 255)
            astronauta_contraste[i, j, c] = nuevo_valor

# Mostrar la imagen original y la modificada
plt.figure(figsize=(10, 5))

# Imagen original
plt.subplot(1, 2, 1)
plt.imshow(astronauta)
plt.title("Imagen Original")

# Imagen con contraste ajustado
plt.subplot(1, 2, 2)
plt.imshow(astronauta_contraste)
plt.title(f"Imagen con Contraste x{factor_contraste}")

plt.show()
````
