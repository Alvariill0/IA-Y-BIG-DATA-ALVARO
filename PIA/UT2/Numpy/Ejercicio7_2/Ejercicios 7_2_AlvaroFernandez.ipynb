## Encontrando patrones en una imagen en blanco y negro

Dada la imagen que os he dejado en el aula virtual, que tiene repartidos simbolos "más" (+) de forma aleatoria:

1. Carga la imagen
2. Construye una matriz de 3x3 que represente un símbolo "+" como los de la imagen
3. Programa una función que encuentre dicho patrón en la imagen
4. Ve guardando las posiciones (coordenadas) donde están los "mases" en otra matrix y representa los datos con matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Cargar la imagen
img = plt.imread('7_2_imagen.png')

# 2. Construye una matriz 3x3 que represente un simbolo "+"
simbolo = np.array([[1, 0, 1],
                    [0, 0, 0],
                    [1, 0, 1]])

plt.figure()
plt.imshow(simbolo, cmap='gray')
plt.title('Simbolo "+"')

#3. Programa una función que encuentre dicho patrón en la imagen
def encontrar_simbolo(imagen, simbolo):
    filas_img, cols_img = imagen.shape[:2]
    filas_sim, cols_sim = simbolo.shape
    posiciones = []
    # trabajamos sobre el primer canal si hay varios
    canal = imagen[..., 0] if imagen.ndim == 3 else imagen
    for i in range(filas_img - filas_sim + 1):
        for j in range(cols_img - cols_sim + 1):
            if np.array_equal(canal[i:i+filas_sim, j:j+cols_sim], simbolo):
                posiciones.append((i, j))
    return posiciones

# Imprime las posiciones encontradas
posiciones = encontrar_simbolo(img, simbolo)
print(f'Símbolos "+" encontrados: {len(posiciones)}')
for p in posiciones:
    print(p)

# 4. Guardar las coordenadas en una lista y representarlas con matplotlib
plt.figure()
plt.imshow(img, cmap='gray')
if posiciones:
    ys = [p[0] + 1 for p in posiciones]  # +1 para que el marcador quede centrado
    xs = [p[1] + 1 for p in posiciones]
    plt.scatter(xs, ys, c='red', s=50, edgecolor='black')
plt.title('Imagen con símbolos "+" marcados')
plt.show()
```
