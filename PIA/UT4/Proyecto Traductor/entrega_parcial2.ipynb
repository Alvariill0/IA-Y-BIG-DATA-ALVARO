```python
from dotenv import load_dotenv
import os
import requests
import json
import azure.cognitiveservices.speech as speechsdk
from azure.ai.translation.text.models import InputTextItem


load_dotenv()

KEY_TRANSLATOR = os.getenv("TRAD_KEY1")
TRAD_ENDPOINT = os.getenv("ENDPOINT_TRAD")
KEY_SPEECH = os.getenv("SPEECH_KEY1")
SPEECH_ENDPOINT = os.getenv("ENDPOINT_SPEECH")
REGION = os.getenv("REGION")
```

#### 1. Identifica funciones necesarias con API

```python
## 1. FUNCIONES NECESARIAS con API REST (Azure Translator)

def listar_idiomas(endpoint, key, region):
    """
        Lista todos los idiomas disponibles.
    """
    url = f"{endpoint}/languages?api-version=3.0&scope=translation"
    headers = {
        "Ocp-Apim-Subscription-Key": key,
        "Ocp-Apim-Subscription-Region": region
    }
    
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        return response.json()
    return None

def detectar_idioma(texto, endpoint, key, region):
    """
        Detecta idioma de un texto.
    """
    url = f"{endpoint}/detect?api-version=3.0"
    headers = {
        "Ocp-Apim-Subscription-Key": key,
        "Ocp-Apim-Subscription-Region": region,
        "Content-Type": "application/json"
    }
    body = [{"Text": texto}]
    
    response = requests.post(url, headers=headers, json=body)
    if response.status_code == 200:
        return response.json()[0]["language"]
    return None

def traducir(texto, idioma_origen, idioma_destino, endpoint, key, region):
    """
        Traduce texto.
    """
    url = f"{endpoint}/translate?api-version=3.0&from={idioma_origen}&to={idioma_destino}"
    headers = {
        "Ocp-Apim-Subscription-Key": key,
        "Ocp-Apim-Subscription-Region": region,
        "Content-Type": "application/json"
    }
    body = [{"Text": texto}]
    
    response = requests.post(url, headers=headers, json=body)
    if response.status_code == 200:
        # Si la traduccion es del mismo idioma
        if idioma_origen == idioma_destino:
            print("El idioma de origen y destino es el mismo. No se realizará traducción.")
        else:
            return response.json()[0]["translations"][0]["text"]
    return None

```

#### 2. Realiza una prueba con cada una de ellas

```python
# Prueba funciones API.
# - Detectar idioma
texto = "Hola que tal como estas"
idioma = detectar_idioma(texto, TRAD_ENDPOINT, KEY_TRANSLATOR, REGION)
print(f"Idioma detectado: {idioma}")

# - Traducir
traduccion = traducir(texto, idioma, "es", TRAD_ENDPOINT, KEY_TRANSLATOR, REGION)
if traduccion:
    print(f"Traducción: {traduccion}")

# - Listar todos los idiomas
idiomas = listar_idiomas(TRAD_ENDPOINT, KEY_TRANSLATOR, REGION)
print("Idiomas disponibles:")
for codigo, nombre in idiomas["translation"].items():
    print(f"{codigo}: {nombre['name']}")
```
