#### 1.	Haciendo uso del fichero stock.csv realiza un programa que abra el fichero, lea sus datos y los almacene en una lista de diccionarios. Después imprime en pantalla el stock de cada material. Prueba a hacerlo con y sin usar la estructura “with” de acceso a fichero.


```python
# Abrir el csv y leer su contenido
with open('stock.csv', 'r') as f1:
    # Obtener cabeceras, crear diccionarios con nombre de la cabecera:
    cabeceras = f1.readline().strip().split(',')
    #Crear diccionarios
    for cabecera in cabeceras:
        cabecera = {'Nombre': cabecera, 'Stock': 0}

    #Leer resto de lineas, modificar el stock de cada producto
    contenido = f1.readlines()
    for linea in contenido:
        valores = linea.strip().split(',')
        diccionarios = {cabeceras[i]: valores[i] for i in range(len(cabeceras))}
        print(diccionarios)
    #cerrar
f1.close()
```

#### 2.	Haciendo uso del fichero stock.csv realiza un programa que abra el fichero, lea los datos y almacene en un diccionario la cantidad acumulada de cada uno de los materiales (arandelas, tornillos y tuercas). Después pregunta al usuario de qué material quiere conocer el stock y muéstralo por pantalla. 
- Añade control de excepciones en tu código. Razonamos dónde es necesario.



```python
# Diccionario stock:
stock = {'tornillos': 0, 'tuercas': 0, 'arandelas': 0}
try:
    with open('stock.csv', 'r') as fich:
        next(fich)  # Saltar la cabecera
        for linea in fich:
            stock['tornillos'] += int(linea.split(',')[0])
            stock['tuercas'] += int(linea.split(',')[1])
            stock['arandelas'] += int(linea.split(',')[2])
        
except FileNotFoundError:
    print("Error:  fichero no encontrado.")
else:
    material = input('Qué material quieres consultar (tornillos/tuercas/arandelas) ').strip().lower()
    if material == 'tornillos':
        print(f"Cantidad de tornillos: {stock['tornillos']}")
    elif material == 'tuercas':
        print(f"Cantidad de tuercas: {stock['tuercas']}")
    elif material == 'arandelas':
        print(f"Cantidad de arandelas: {stock['arandelas']}")
    else:
        print("No existe ese material.")
finally:
    fich.close()
    print(stock)
```
