#### 1.	Escribe un programa que pida un número por teclado y que cree un diccionario cuyas claves sean desde el número 1 hasta el número indicado, y los valores sean los cuadrados de las claves.


````python
diccionario1 = {}

long = int(input("Ingrese la cantidad de elementos del diccionario: "))
for i in range(long):
    diccionario1[i+1] = (i+1)**2

diccionario1
````

#### 2.	Escribe un programa que lea una cadena y devuelva un diccionario con la cantidad de apariciones de cada carácter en la cadena.


````python
cadena = input("Escribe una cadena: ")
diccionario2 = {}
for c in cadena:
    if c in diccionario2:
        diccionario2[c] += 1
    else:
        diccionario2[c] = 1
diccionario2
````

#### 3.	Vamos a crear un programa en donde vamos a declarar un diccionario para guardar los precios de 4 frutas a nuestra elección. Posteriormente el programa pedirá el nombre de una fruta y la cantidad que se ha vendido y nos mostrará el precio final de la fruta a partir de los datos guardados en el diccionario. Si la fruta no existe nos dará un error.  Tras cada consulta el programa nos preguntará si queremos hacer otra consulta.


````python
diccionario3 = {
    'platano' : 1,
    'manzana' : 2,
    'pera' : 3,
    'naranja' : 4
}
while True:
    fruta = input("Ingrese el nombre de una fruta: ").lower()
    cant = int(input("Cuanta cantidad se ha vendido de esa fruta?: "))

    if fruta in diccionario3:
        print("El valor total de la compra serían:",diccionario3[fruta]*cant,"€.")
    else:
        print("No hay datos de esa fruta.")
    salir = input("Quiere hacer otra consulta? (s/n)").lower()
    if salir == 'n':
        print("Chao pescao")
        break
    
````

#### 4.	Dado un número entero introducido por teclado, guarda sus divisores en un conjunto y muéstralo.

````python
num = int(input("Introduce un num entero: "))
conj = set(())
for i in range(num):
    if num % (i+1) == 0:
        conj.add(i+1)

conj
````

#### 5.	Dada una frase introducida por teclado, guarda en un conjunto la primera letra de cada palabra

````python
frase = input("Escribe una frase: ")
palabras = frase.split()
conj = set(())
for palabra in palabras:
    conj.add(palabra[0])

conj
````

#### 6.	Dada una frase proporcionada por el usuario, crea una lista de tuplas indicando palabra, longitud de cada palabra, letra inicial y posición que ocupan dentro de la frase.


````python
frase = input("Escribe una frase")
lista = []
palabras = frase.split()
for palabra in palabras:
    #palabra, longitud, primera letra, posicion en la frase
    lista.append((palabra,len(palabra),palabra[0], palabras.index(palabra)))

lista
````

#### 8.	Implementa un programa que nos permita guardar los nombres de los alumnos de una clase y las notas que han obtenido. Cada alumno puede tener distinta cantidad de notas. Guarda la información en un diccionario cuyas claves serán los nombres de los alumnos y los valores serán una lista con las notas de cada alumno. El programa pedirá el número de alumnos que vamos a introducir, pedirá su nombre e irá pidiendo sus notas hasta que introduzcamos un número negativo. 
El programa realizará después dos tareas: mostrará la lista de alumnos y la nota media obtenida por cada uno de ellos y pedirá el nombre de un alumno y mostrará sus notas y la media.

Nota: si se introduce el nombre de un alumno que ya existe el programa nos dará un error.

````python
alumnos = {
    'pepe' : [5,5,7,6],
    'maria' : [7,7,3,9,2],
    'juan' : [2,2,2],
    'nerea' : [7,7,7,7]
}
long = int(input("Cuántos alumnos quieres introducir?:"))

if long != 0:
    for i in range(long):
        notas = []
        nombre = input("Nombre del alumno").lower()
        if nombre in alumnos.keys():
            print("Ese nombre ya existe en la clase")
        else:
            alumnos[nombre] = notas
            while True:
                nota = int(input("Introduce una nota del alumno (negativo para salir)"))
                if nota >= 0:
                    notas.append(nota)
                else:
                    break

consulta = input("Nombre del alumno que quieres ver:").lower()
if consulta in alumnos.keys():
    print(f"{alumnos[consulta]}")
else:
    print("Ese alumno no existe")

print("-- LISTA DE ALUMNOS Y MEDIAS")
for clave, valor in alumnos.items():
    print(f"{clave}, Nota media: {sum(valor) / len(valor)}")
````

#### 9.	Haz que el usuario introduzca palabras hasta que introduzca una palabra vacía. Guarda todas las palabras en una tupla y muestra la primera y la última introducidas haciendo uso de unpacking. 
PISTA: Para guardar los elementos de uno en uno vas a tener que utilizar un tipo de dato que no es tupla y luego transformarlo a tupla.

Usamos unpacking para obtener el primer y último valor, y descartamos el resto

```
primero, *_, ultimo = tupla
print(primero)  # Salida: 10
print(ultimo)   # Salida: 50
```

````python
lista_palabras = []
while True:
    palabra = input("Introduce una palabra (vacía para salir): ")
    if palabra == "":
        break
    else: 
        lista_palabras.append(palabra)
tupla = tuple(lista_palabras)
primero, *_, ultimo = tupla
print(f"Primero: {primero}")  
print(f"Último: {ultimo}")  
````


#### 10.	Dada una lista de palabras, crea otra lista del mismo tamaño que guarde la longitud de cada palabra. Usa la función zip() para crear un diccionario con claves las palabras y valores, su longitud.

````python
palabras = ["manzana", "platano", "pera", "naranja", "kiwi"]
long = []
diccionario10 = {}
for palabra in palabras:
    long.append(len(palabra))

for palabra, longitud in zip(palabras,long):
    diccionario10[palabra] = longitud

diccionario10
````
