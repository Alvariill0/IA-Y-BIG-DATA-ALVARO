### **SLICING**: 

#### 1. Slicing básico
-   Dada la cadena `mensaje = 'Python es genial'`, obtén los primeros 6 caracteres 


```python
cadena = 'Python es genial'
print(cadena[:6])
```

#### 2. Slicing con salto
-   Dada la cadena `alfabeto = 'abcdefghijklmnopqrstuvwxyz'`, obtén los caracteres pares de la cadena.


```python
alfabeto = 'abcdefghijklmnopqrstuvwxyz'
print(alfabeto[::2])
```

#### 3. Slicing inverso
-   Dada la cadena `frase = 'Invertir una cadena es sencillo'`, invierte el orden de los caracteres.


```python
frase = 'Invertir una cadena es sencillo'
print(frase[::-1])
```

#### 4. Slicing con índices negativos
-   Dada la cadena `oracion = 'La tecnología es el futuro'`, obtén los últimos 8 caracteres usando índices negativos.


```python
oracion = 'La tecnología es el futuro'
print(oracion[-8:])
```

#### 5. Slicing de palabras
-   Dada la cadena `texto = 'Aprender Python paso a paso'`, obtén la palabra 'Python' utilizando slicing.


```python
texto = 'Aprender Python paso a paso'
print(texto[9:15])
```

#### 6. Slicing de subcadena especifica
-   Dada la cadena `eslogan = 'El poder de la programación'`, obtén los caracteres desde el índice 3 al 14.


```python
alfabeto = 'abcdefghijklmnopqrstuvwxyz'
print(alfabeto[::2])
```

#### 7. Slicing con paso inverso
-   Dada la cadena `cadena = 'abcdefghijk'`, obtén los caracteres de la cadena en orden inverso, pero saltando uno.


```python
cadena = 'abcdefghijk'
print(cadena[::-2])
```

#### 8. Slicing de mitad de cadena
-   Dada la cadena `mensaje = 'Estamos aprendiendo slicing'`, obtén la primera mitad de los caracteres de la cadena.

```python
mensaje = 'Estamos aprendiendo slicing'
mitad = len(mensaje) // 2
print(mensaje[:mitad])
```

#### 9. Slicing de frase truncada
-   Dada la cadena `frase = 'El conocimiento es poder'`, obtén solo la frase 'El conocimiento'.


```python
frase = 'El conocimiento es poder'
print(frase[0:15])
```

#### 10. Slicing con paso personalizado
-   Dada la cadena `palabras = 'abcdefghijklmnopqrstuvwxz'`, obtén los caracteres en posiciones múltiplos de 3 (cada tercer carácter).


```python
palabras = 'abcdefghijklmnopqrstuvwxz'
print(palabras[::-3])
```
