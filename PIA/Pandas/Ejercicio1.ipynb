# EJERCICIO 1 PANDAS


```python
import pandas as pd

df = pd.read_csv('./datos_usuarios.csv')
```

```python
# 1. Nombre y correo de todos los usuarios
df[['name', 'email']]
```

```python
# 2. Nombre de usuarios Femenino
df[df['gender'] == 'Femenino']['name']
```

```python
# 3. Masculino y edad > 50
df[(df['gender'] == 'Masculino') & (df['age'] > 50)]['name']
```

```python
# 4. Edad > 20
print(df[df['age'] > 20])
```

```python
# 5. Username y email, edad 50-60
print(df[(df['age'] >= 50) & (df['age'] <= 60)][['username', 'email']])
```

```python
# 6. Username con email empezando "e"
print("\n6. Email empieza con 'e':")
print(df[df['email'].str.startswith('e')]['username'])

# 7. Username Alemania/Finlandia/Canadá
print("\n7. Países específicos:")
paises_especificos = ['Alemania', 'Finlandia', 'Canada']
print(df[df['country'].isin(paises_especificos)]['username'])

# 8. Femenino de Alemania (nombre y email)
print("\n8. Mujeres Alemania:")
print(df[(df['gender'] == 'Femenino') & (df['country'] == 'Alemania')][['name', 'email']])

# 9. Promedio edad mujeres Canadá >20
print("\n9. Promedio edad mujeres Canadá >20:")
promedio = df[(df['gender'] == 'Femenino') & (df['country'] == 'Canada') & (df['age'] > 20)]['age'].mean()
print(promedio)

# 10. Cantidad usuarios Finlandia
print("\n10. Usuarios Finlandia:")
print(df[df['country'] == 'Finlandia'].shape[0])
```

```python
# 11. Groupby género
print("\n11. Cantidad por género:")
print(df.groupby('gender').size())

# 12. País con más mujeres (sort_values)
print("\n12. País con más mujeres:")
mujeres_por_pais = df[df['gender'] == 'Femenino'].groupby('country').size().sort_values(ascending=False)
print(mujeres_por_pais.head(1))

# 13. 3 países con más usuarios
print("\n13. Top 3 países:")
usuarios_por_pais = df['country'].value_counts().head(3)
print(usuarios_por_pais)
```

```python
# 14. Edad 10-20 o 40-70
print("\n14. Rangos edad:")
print(df[(df['age'].between(10,20)) | (df['age'].between(40,70))]['name'])

# 15. Usuario más joven Canadá
print("\n15. Más joven Canadá:")
mas_joven_canada = df[(df['country'] == 'Canada')].nsmallest(1, 'age')[['name', 'email']]
print(mas_joven_canada)

# 16. Usuario más viejo Canadá
print("\n16. Más viejo Canadá:")
mas_viejo_canada = df[(df['country'] == 'Canada')].nlargest(1, 'age')[['name', 'email']]
print(mas_viejo_canada)

# 17. 3 países con menos usuarios
print("\n17. 3 países menos usuarios:")
print(df['country'].value_counts().tail(3))

# 18. País con más usuarios >50
print("\n18. País más usuarios >50:")
mayores_50 = df[df['age'] > 50]['country'].value_counts().head(1)
print(mayores_50)
```

```python
# 19. País con mayor promedio edad
print("\n19. País mayor promedio edad:")
promedio_edad_pais = df.groupby('country')['age'].mean().sort_values(ascending=False).head(1)
print(promedio_edad_pais)

# 20. País con más hombres
print("\n20. País más hombres:")
hombres_por_pais = df[df['gender'] == 'Masculino'].groupby('country').size().sort_values(ascending=False).head(1)
print(hombres_por_pais)

# 21. Códigos postales México
print("\n21. Postcodes México:")
print(df[df['country'] == 'Mexico']['postcode'].tolist())

# 22. Edad más repetida
print("\n22. Edad más repetida:")
edad_max = df['age'].value_counts().head(1)
print(edad_max)

# 23. Edad menos repetida
print("\n23. Edad menos repetida:")
edad_min = df['age'].value_counts().tail(1)
print(edad_min)
```
