## 1. Introducción

En  Computación  y  Ciencia  de datos,  *Pandas*  es  una  biblioteca de  software  escrita  como  extensión  de NumPy para manipular, analizar, limpiar y explorar datos para el lenguaje de programación Python.

Pandas nos permite analizar big data y sacar conclusiones basadas en teorías estadísticas.

Por tratarse de una librería especializada en el procesamiento de datos estructurados, muchas de las operaciones implementadas en Python siguen la lógica de funcionamiento de NumPy, la librería para el procesamiento de arrays.

Además, con ayuda de Pandas podemos abordar problemas de Análisis de Datos y Análisis Exploratorio de Datos en nuestros proyectos de Ciencia de Datos y Machine Learning. 
Esto muchas veces requiere el uso de herramientas de visualización de datos que en el fondo usan librerías como Matplotlib

Podemos: limpiar  conjuntos  de  datos  desordenados  y  hacerlos  legibles  y  relevantes.  También se puede eliminar filas que no son relevantes o que contienen valores incorrectos, como valores vacíos o NULL.

El nombre "Pandas" hace referencia tanto a "Datos de panel" como a "Análisis de datos de Python".

Pandas resulta fundamental en el ciclo de vida de un proyecto de Ciencia de Datos y o de Machine Learning en las siguientes fases:

- En la limpieza y exploración de los datos
- En el manejo de datos faltantes o de valores extremos
- En el análisis (descriptivo, explicativo, predictivo o prescriptivo) de los datos
- En el pre-procesamiento de los datos y la extracción de características
- En la comunicación de resultados

Vale la pena aclarar que podemos usar Pandas en un proyecto de Machine Learning siempre y cuando los datos que estemos procesando sean datos estructurados.

Además de permitir manipular únicamente datos estructurados, existe otra situación en la cual es recomendable el uso de Pandas: cuando tenemos “datos pequeños” es decir cuando nuestro set de datos se puede almacenar en memoria en un sólo ordenador.

Cuando tenemos datasets gigantescos, que superan la capacidad de memoria de nuestro computador, es recomendable usar otras librerías como por ejemplo PySpark, Dask o Polars.

Sin embargo, es importante resaltar que estas librerías usan una sintaxis muy similar a la usada por Pandas para el procesamiento de datasets pequeños, así que el hecho de aprender Pandas nos permitirá reducir posteriormente la curva de aprendizaje de estas otras librerías en caso de que sea necesario aprenderlas.

### a.	Definiendo Series

En el caso de strings, al no ser de tipo primitivo, aparece como de tipo object.
 
En caso de que nuestros elementos que conforman el índice y el numero de elementos de contenido sean diferentes tendremos un error de ejecución.

Si desconociéramos algún valor, podríamos rellenar los elementos vacíos con el elemento None.


En cambio, si en vez de None utilizasemos un string, el tipo dejaría de ser primitivo para ser tipo  object, como demuestra el siguiente ejemplo.

### C. Trabajando con aleatorios y grandes series

En machine learning normalmente se trabaja con grandes cantidades de datos. Una manera simple de conseguirlo es generándolos aleatoriamente.

Además, generar series aleatorias es muy práctico, ya que nos será útil para simular datos de prueba.
Para generar los números aleatorios, además de Pandas, necesitaremos de la librería Numpy.
Importamos Numpy si no lo hemos hecho ya.

¡OJO! Funciones como print() dejan de ser prácticas cuando las series alcanzan cierto tamaño, ya que inundarían la pantalla con demasiados datos.

Es por ello por lo que se hace necesario el uso de funciones alternativas como head() o tail(). Por defecto muestran los 5 primero o ultimos valores

````python
import numpy as np

# Creamos una Lista compuestas por 10.800 números aleatorios comprendidos entre 10 y 99
serieNumerosAleatorios = pd.Series(np.random.randint(10,99,10008))

serieNumerosAleatorios.head()

serieNumerosAleatorios.tail()

# Mostramos Los 10 primeros elementos
serieNumerosAleatorios.head(10)

#Mostramos Los 10 últimos elementos
serieNumerosAleatorios.tail(10)
````

### d. Aplicando operaciones sobre una Serie

Dado que las series se pueden recorrer como cualquier otra estructura iterativa, siempre podemos implementar las operaciones que necesitemos por nosotros mismos.

Si queremos sumar todos los elementos de una serie, podríamos:

````python
# Sumatorio de forma manual
total = 0
for elemento in serieNumerosAleatorios:
    total += elemento
print(total)


# Sumatorio utilizando funciones de Numpy

sumatorio = np.sum(serieNumerosAleatorios)
print(sumatorio)
````

Si tomamos por ejemplo la media, tendríamos:

````python
# Media de forma manual
total = 0
for elemento in serieNumerosAleatorios:
    total += elemento
print(total/len(serieNumerosAleatorios))


# Media utilizando funciones de Numpy

media = np.mean(serieNumerosAleatorios)
print(media)
````

Ocurre lo mismo con otras operaciones, como por ejemplo añadir o restar un valor constante a todos los
elementos de la serie. implementándolo por nosotros mismos quedaría del siguiente modo

````python
serieNumerosAleatorios = pd.Series(np.random.randint(10,99,100))
serieNumerosAleatoriosCopia = serieNumerosAleatorios.copy()


#Restamos 10 a todos Los elementos
for indice, elemento in serieNumerosAleatorios.items():
    serieNumerosAleatorios[indice] = elemento - 10
# Comprobamos como La media se ha reducido en 10
print(np.mean(serieNumerosAleatorios))


## Mientras que esa misma operación se podría realizar de la siguiente manera

serieNumerosAleatoriosCopia -= 10
# Comprobamos como La media se ha incrementado en 10, volviendo a su estado original
print(np.mean(serieNumerosAleatoriosCopia))
````



La potencia de la serie nos permite aplicar la operación directamente:

````python
serieNumerosAleatoriosCopia -= 10
# Comprobamos como La media se ha incrementado en 10, volviendo a su estado original
print(np.mean(serieNumerosAleatoriosCopia))
````

A esto se le conoce como VECTORIZACIÓN y podremos implementarlo para:

- Operaciones matemáticas básicas
- Operaciones entre series (serie1 + serie2)
- Operaciones de numpy
- Cadenas de caracteres. A través del espacio de nombres .str podemos aplicar funciones string a una serie

Internamente se utilizan funciones avanzadas de numpy, esto hace que sea más eficiente que recorrelo con un bucle.

````python
import pandas as pd

# DataFrame de ejemplo con una columna de strings
df_ejemplo = pd.DataFrame({
    'ID': [1, 2, 3],
    'Nombres': ['ESPAÑA', 'Francia', 'Alemania'],
    'Valores': [100, 200, 300]
})

df_ejemplo['Nombres_min'] = df_ejemplo['Nombres'].str.lower()

print(df_ejemplo)
````

### ¡¡OJO!!: Un índice es un identificador simple o compuesto de una fila, no es una identificador único como una clave primaria. 

- Indice simple. ```.loc``` devolverá todas las filas que coincidan con ese indice
- Indice multinivel: ```.loc``` devolverá las filas que coincidan con los niveles especificados

````python

data_simple = {
    'Población': [500000, 600000, 1000000, 700000],
    'País': ['España', 'España', 'Francia', 'Francia']
}
ciudades = ['Madrid', 'Madrid', 'París', 'París']

# Crear DataFrame con índice duplicado
df_simple = pd.DataFrame(data_simple, index=ciudades)

# Mostrar el DataFrame
print("DataFrame con índice simple duplicado:\n", df_simple)

# Acceder a todas las filas con índice 'Madrid'
print("\nFilas con índice 'Madrid':\n", df_simple.loc['Madrid'])
````

````python
# Crear datos
data_multinivel = {
    'Población': [500000, 300000, 1000000, 700000],
    'Idioma': ['Español', 'Español', 'Francés', 'Francés']
}
indice_multinivel = [('Madrid', 'Centro'), ('Madrid', 'Sur'), ('París', 'Centro'), ('París', 'Centro')]

# Crear DataFrame con índice multinivel
df_multinivel = pd.DataFrame(data_multinivel, index=pd.MultiIndex.from_tuples(indice_multinivel, names=['Provincia', 'Ciudad']))

# Mostrar el DataFrame
print("DataFrame con índice multinivel duplicado:\n", df_multinivel)

# Acceder a todas las filas con el índice ('París', 'Centro')
print("\nFilas con índice ('París', 'Centro'):\n", df_multinivel.loc[('París', 'Centro')])
````

### - Agrupar datos 

La función groupby() permite agrupar datos en función de una o más columnas 

````python

data = {'Producto': ['A', 'A', 'B', 'B', 'C', 'C'],
        'Tienda': ['T1', 'T2', 'T1', 'T2', 'T1', 'T2'],
        'Ventas': [100, 150, 200, 250, 300, 350]}

df = pd.DataFrame(data)

# Calcular la media de ventas para cada combinación de Producto y Tienda
resultado_groupby = df.groupby(['Producto', 'Tienda'])['Ventas'].mean()
print(resultado_groupby)
````

### - Crear pivot tables con pivot_table()
Para resumir y analizar grandes cantidades de datos.
Similar a group_by pero este permite varios niveles (Piensa en las pivot tables de excel)

````python

data = {
    'Producto': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Tienda': ['T1', 'T2', 'T1', 'T2', 'T1', 'T2'],
    'Ventas': [100, 150, 200, 250, 300, 350]
}

df = pd.DataFrame(data)

# Crear una tabla dinámica con el promedio de ventas por tienda y producto
tabla_pivot = df.pivot_table(values='Ventas', index='Producto', columns='Tienda', aggfunc='mean')
print(tabla_pivot)
````

#### VI. Mezclar dataframes (merge)

Permite integrar filas de dos DataFrames que contienen información en común en una o varias columnas o índices que se conocen como clave.

Se utiliza el método **merge** que puede tener los siguientes tipos:

- inner: intersección de conjuntos. Filas cuyos valores clave están en los dos dataframes
- outer: union de conjuntos. Las filas que no pueden emparejarse se completan con Nan
- left: se descartan las filas del segundo dataframe que no puedan emparejarse con alguna del primero a través de la clave
- right: se descartan las filas del primer dataframe que no puedan emparejarse con alguna del primero a través de la clave


````python
import pandas as pd
df1 = pd.DataFrame({"Nombre":["Carmen", "Luis", "María"],  "Sexo":["Mujer", "Hombre", "Mujer"]})
df2 = pd.DataFrame({"Nombre":["María", "Pedro", "Luis"], "Edad":[25, 30, 18]})
df = pd.merge(df1, df2, on="Nombre", how="left")
print(df)
````

## 7.	Trabajar con Bases de Datos

Pandas nos permite interactuar con una base de datos a través del método **to_sql**


El método to_sql() tiene varios parámetros que ajustan la funcionalidad y los datos que se envían a la base de datos. Estos incluyen:

- name: nombre de la tabla SQL (obligatorio)
- con: conexión proporcionada por sqlite o sqlalchemy (obligatorio)
- schema: esquema de la base de datos SQL
- if_exists: qué hacer si la tabla ya existe, ya sea 'fail' (fallar), 'replace' (reemplazar) o 'append' (agregar)
- index: si se debe escribir la columna de índice del DataFrame en la tabla o no
- index_label: el nombre para dar a la columna de índice si se escribe en la tabla
- chunksize: número de filas que se escribirán a la vez
- dtype: diccionario de todas las columnas y sus tipos de datos
- method: especificar la cláusula de inserción SQL

De forma similar la función read_sql() permite obtener datos de una o varias tablas, pasándole como parámetro la query
a realizar

Necesitaremos instalar **pip install pandas sqlalchemy** para la gestión con la base de datos y **pip install pymysql**


La función **create_engine** poporciona una conexión a una BD pasada en una URL


````python

import pandas as pd
import sqlalchemy as sq
from sqlalchemy import create_engine


# Reemplaza datos de conexión
engine = create_engine('mysql+pymysql://root:toor@localhost/temperaturas')  

datos = pd.read_html("https://es.wikipedia.org/wiki/Territorio_dependiente")

df = datos[0] #Generar dataframe y trabajar

df.to_sql('paises_dependientes', con=engine, if_exists='replace', index=False)
````
