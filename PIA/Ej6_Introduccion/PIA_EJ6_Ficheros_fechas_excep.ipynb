#### 1.	Dado un fichero ‘frutas.txt’ con una palabra por línea, crear un fichero de salida con las frutas ordenadas.


```python
with open('frutas.txt', 'r') as file:
    frutas = file.readlines()
    #Crear fichero salida con las frutas ordenadas alfabeticamente
    lista = []
    for fruta in frutas:
        lista.append(fruta.strip())
    lista.sort()
    with open('frutas_ordenadas.txt', 'w') as salida:
        for fruta in lista:
            salida.write(fruta + '\n')
```

#### 3.	Abre el archivo ‘personas.txt’ y añade datos de nuevas personas (nombre y edad) al final. Hazlo preguntando al usuario hasta que este diga que no quiere seguir.

```python
while True:
    #Preguntar al usuario nombre y edad.
    nombre = input("Nombre de la persona a añadir: ")
    edad = input("Edad de la persona a añadir: ")
    with open('personas.txt', 'a') as personas:
        personas.write(f"{nombre}, {edad}\n")

    res = input("¿Deseas añadir otra persona? (s/n): ")
    if res.lower() != 's':
        break
```

#### 4.	Comprobar si un archivo llamado ‘alumnos.txt’ existe y si no, crearlo. Pide al usuario Nombre de alumno, módulo y nota para guardar en el fichero una línea por alumno, con los datos separados por comas.

```python
# Comprobar que existe el archivo.
import os
if os.path.exists('alumnos.txt'):
    print("El archivo 'alumnos.txt' ya existe. No se va a crear uno nuevo.")
else:
    alumnos = open('alumnos.txt', 'a')
    while True:
        nombre = input("Nombre del alumno: ")
        modulo = input("Módulo del alumno: ")
        nota = input("Nota del alumno: ")
        alumnos.write(f"{nombre}, {modulo}, {nota}\n")

        res = input("¿Deseas añadir otro alumno? (s/n): ")
        if res.lower() != 's':
            break
    
```

#### 5.	Abre el fichero anterior, que contiene notas de los alumnos y recórrelo para calcular la media de las notas de cada módulo.

```python
with open('alumnos.txt', 'r') as alumnos:
    contenido = alumnos.read()
    # Buscar solo las notas y calcular media
    lineas = contenido.splitlines()
    notas = []
    for linea in lineas:
        partes = linea.split(',')
        if len(partes) == 3:
                nota = float(partes[2].strip())
                notas.append(nota)
    if notas:
        media = sum(notas) / len(notas)
        print(f"La nota media de los alumnos es: {media:.2f}")
    else:
        print("No hay notas registradas.")
```

#### 6.	Dados dos ficheros con una palabra por línea, implementa una función que reciba los nombres de los ficheros como parámetro y devuelva un fichero de salida que sea la concatenación de las correspondientes líneas de cada fichero de entrada.


```python
def concatenacion(fich1, fich2):
    #Crear fichero
    try:
        with open(fich1, 'r') as f1, open(fich2, 'r') as f2, open('salida_concatenada.txt', 'w') as salida:
            lineas1 = f1.readlines()
            lineas2 = f2.readlines()
            max_len = max(len(lineas1), len(lineas2))
            for i in range(max_len):
                if i < len(lineas1):
                    salida.write(lineas1[i].strip() + '\n')
                if i < len(lineas2):
                    salida.write(lineas2[i].strip() + '\n')
    except FileNotFoundError:
        print("Uno o varios ficheros no existen.")
    return salida

concatenacion('frutas.txt', 'cosas.txt')
```
